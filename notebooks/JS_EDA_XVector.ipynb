```python
import torch

print("Number of GPU: ", torch.cuda.device_count())
print("GPU Name: ", torch.cuda.get_device_name())


device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print('Using device:', device)
```

```python
import pandas as pd
import numpy as np
import mne
import os
import matplotlib.pyplot as plt
```

```python
from eeg_lib.commons.constant import DATASETS_FOLDER
from eeg_lib.data.data_loader.EEGDataExtractor import EEGDataExtractor
```

```python
DATA_DIR = f"{DATASETS_FOLDER}/Kolory/"

extractor = EEGDataExtractor(data_dir=DATA_DIR)
eeg_df, participants_info = extractor.extract_dataframe()
```

```python
import torch.optim as optim
from torch.utils.data import DataLoader

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.manifold import TSNE
from sklearn.preprocessing import MinMaxScaler

import matplotlib.patches as mpatches
from matplotlib.colors import ListedColormap
```

```python
from eeg_lib.data.data_loader.TDNNFeatures import extract_features
from eeg_lib.data.TDNNDataset import TDNNDataset
from eeg_lib.models.verification.XVector import XVectorEmbeddingModel
from eeg_lib.losses.ProxyNCALoss import ProxyNCALoss
from eeg_lib.utils.visualisations import plot_tsne
from eeg_lib.models.similarity.centroids import SimilarityCentroidsVerifier
```

## Hyperparameters

```python
batch_size = 16
num_epochs = 20
learning_rate = 0.001
```

## Feature extraction

```python
extracted_features = []
for epoch in eeg_df["epoch"]:
    extracted_features.append(extract_features(epoch, frame_length=50).T)
```

### Standardization

```python
extracted_features = np.array(extracted_features)
```

```python
mean = np.mean(extracted_features, axis=2, keepdims=True)
std = np.std(extracted_features, axis=2, keepdims=True)
```

```python
extracted_normalized = (extracted_features - mean) / std
```

### train test split and model creation

```python
y = eeg_df["participant_id"].values
```

```python
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(extracted_normalized,y_encoded, test_size=0.2, random_state=42)
```

```python
dataset = TDNNDataset(X_train, y_train)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

model = XVectorEmbeddingModel(input_features=extracted_normalized.shape[1],
                              layer1_filt=128,
                              layer2_filt=128,
                              layer3_filt=128,
                              layer4_filt=128,
                              layer5_filt=1000,
                              dropout1=0,
                              dropout2=0,
                              dropout3=0,
                              dropout4=0).to(device)
proxy_loss_fn = ProxyNCALoss(num_classes=30, embedding_dim=32,scale=10.0).to(device)
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
```

## Model training

```python
model.train()
for epoch in range(num_epochs):
    epoch_loss = 0
    for data, label in dataloader:
        data = data.to(device)
        label = label.to(device).long()
        optimizer.zero_grad()
        data_embedd = model(data)
        
        loss = proxy_loss_fn(data_embedd, label)
        loss.backward()
        optimizer.step()
        
        epoch_loss += loss.item()
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss}")
```

## Getting the embeddings and visualization

```python
model.to('cpu')
```

```python
embeddings = []
```

```python
with torch.no_grad():
    for epoch in X_train:
        embeddings.append(model(torch.tensor(epoch, dtype=torch.float, requires_grad=False).unsqueeze(0)))
```

```python
test_embeddings = []
```

```python
with torch.no_grad():
    for epoch in X_test:
        test_embeddings.append(model(torch.tensor(epoch, dtype=torch.float, requires_grad=False).unsqueeze(0)))
```

### Visualization

```python
tsne = TSNE(n_components=2, random_state=42)
```

```python
embd = torch.stack(embeddings).reshape((5568,32)).numpy()
```

```python
embd_test = torch.stack(test_embeddings).reshape((1392,32)).numpy()
```

```python
combined_embeddings = np.concatenate((embd, embd_test), axis=0)
```

```python
reduced_combined = tsne.fit_transform(combined_embeddings)
```

```python
train_reduced = reduced_combined[:embd.shape[0]]
test_reduced = reduced_combined[embd.shape[0]:]
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
train_reduced_normalized = scaler.fit_transform(train_reduced)
test_reduced_normalized = scaler.transform(test_reduced)
```

```python
custom_colors = ['#e6194b', '#3cb44b', '#ffe119', '#4363d8', '#f58231', '#911eb4',
                 '#46f0f0', '#f032e6', '#bcf60c', '#fabebe', '#008080', '#e6beff',
                 '#9a6324', '#fffac8', '#800000', '#aaffc3', '#808000', '#ffd8b1',
                 '#000075', '#808080', '#ffffff', '#000000', '#a9a9a9', '#ff69b4',
                 '#b0e0e6', '#32cd32', '#ff4500', '#da70d6', '#ff1493', '#7fffd4']
cmap = ListedColormap(custom_colors[:30])
```

```python
label_encoder = LabelEncoder()
labels_encoded = label_encoder.fit_transform(y_train)
test_labels_encoded = label_encoder.transform(y_test)
```

```python
unique_ids = np.unique(y_train)
handles = []
num_ids = len(unique_ids)
for uid in unique_ids:
    encoded_value = label_encoder.transform([uid])[0]
    color = cmap(encoded_value / num_ids)
    patch = mpatches.Patch(color=color, label=uid)
    handles.append(patch)
```

```python
centroid_verifier = SimilarityCentroidsVerifier(train_reduced_normalized, labels_encoded)
centroids = centroid_verifier.centroids
```

### Visualization

Good on training data

```python
plot_tsne(train_reduced_normalized,
          cmap,
          labels_encoded,
          handles=handles,
          alpha=0.3,
          title="TSNE Visualization training data XVector",
          xlabel="TSNE embedding dimension 1",
          ylabel="TSNE embedding dimension 2",
          centroids=centroids)
```

Not so good on test data

```python
plot_tsne(train_reduced_normalized,
          cmap,
          labels_encoded,
          handles=handles,
          alpha=0.03,
          title="TSNE Visualization test data XVector",
          xlabel="TSNE embedding dimension 1",
          ylabel="TSNE embedding dimension 2",
          centroids=centroids,
          test_embeddings=test_reduced_normalized,
          test_labels=test_labels_encoded)
```
