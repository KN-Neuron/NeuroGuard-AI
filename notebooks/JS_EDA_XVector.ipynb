```python
writer = SummaryWriter(log_dir='runs/TDNN')
```

## Hyperparameters

```python
batch_size = 16
num_epochs = 15
learning_rate = 0.001
```

## Feature extraction

```python
X_train_tmp, X_test_tmp, y_train_tmp, y_test_tmp = split_train_test(eeg_df=eeg_df,
                                                                    test_size=0.2,
                                                                    random_state=42)
```

```python
extracted_X_train = []
for epoch in X_train_tmp:
    extracted_X_train.append(extract_features(epoch, frame_length=50).T)
```

```python
extracted_X_test = []
for epoch in X_test_tmp:
    extracted_X_test.append(extract_features(epoch, frame_length=50).T)
```

```python
le_train = LabelEncoder()
le_train.fit(y_train_tmp)
y_train_encoded = le_train.transform(y_train_tmp)

le_test = LabelEncoder()
le_test.fit(y_test_tmp)
y_test_encoded = le_test.transform(y_test_tmp)
```

### Standardization

```python
X_train = np.array(extracted_X_train)
X_test = np.array(extracted_X_test)
```

```python
mean_train = np.mean(X_train, axis=2, keepdims=True)
std_train = np.std(X_train, axis=2, keepdims=True)
mean_test = np.mean(X_test, axis=2, keepdims=True)
std_test = np.std(X_test, axis=2, keepdims=True)
```

```python
X_train_norm = (X_train - mean_train) / std_train
X_test_norm = (X_test - mean_test) / std_test
```

### train test split and model creation

```python
dataset = TDNNDataset(X_train_norm, y_train_encoded)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

model = XVectorEmbeddingModel(input_features=X_train_norm.shape[1],
                              dropout1=0,
                              dropout2=0,
                              dropout3=0,
                              dropout4=0).to(device)
proxy_loss_fn = ProxyNCALoss(num_classes=30, embedding_dim=32,scale=5.0).to(device)
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
```

## Model training

```python
centroid_verifier = SimilarityCentroidsVerifier()
```

```python
model.train()
for epoch in range(num_epochs):
    epoch_loss = 0
    epoch_acc = 0
    counter = 0
    for data, label in dataloader:
        data = data.to(device)
        label = label.to(device).long()
        optimizer.zero_grad()
        data_embedd = model(data)
        centroid_verifier.update_centroids(data_embedd, label)
        classified = centroid_verifier.classify_batch(data_embedd)
        
        i = 0
        for l in classified:
            counter+=1
            # print(f"classified: {l[0]}\n true label: {label[i]}")
            if l[0] == label[i]:
                epoch_acc += 1
            i += 1
        
        loss = proxy_loss_fn(data_embedd, label)
        loss.backward()
        optimizer.step()
        
        epoch_loss += loss.item()
    avg_loss = epoch_loss / len(dataloader)
    avg_acc  = epoch_acc / (len(dataloader) * batch_size)
    epoch_acc = epoch_acc / (len(dataloader)*batch_size)
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss}")
    print(f"Epoch {epoch+1}/{num_epochs}, Acc: {epoch_acc}")
    writer.add_scalar('Loss/train', avg_loss, epoch)
    writer.add_scalar('Accuracy/train', avg_acc, epoch)
```
