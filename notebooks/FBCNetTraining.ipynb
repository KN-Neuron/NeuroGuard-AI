```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# Define frequency bands (corrected the typo in alpha2)
freq_bands = {
    "delta": (0.5, 4),
    "theta": (4, 8),
    "alpha1": (8, 10),
    "alpha2": (10, 12),
    "beta1": (12, 16),
    "beta2": (16, 20),
    "beta3": (20, 30),
    "gamma1": (30, 40),
    "gamma2": (40, 100),
}

def bandpass_filter(data, lowcut, highcut, fs, order=4):
    """
    Apply a bandpass filter to the signal.
    
    Parameters:
    - data: 1D array of signal data
    - lowcut: low frequency cutoff
    - highcut: high frequency cutoff
    - fs: sampling frequency (Hz)
    - order: filter order
    
    Returns:
    - filtered_data: bandpass filtered signal
    """
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    
    # Create Butterworth bandpass filter
    b, a = signal.butter(order, [low, high], btype='band')
    
    # Apply filter
    filtered_data = signal.filtfilt(b, a, data)
    
    return filtered_data

# Example usage with synthetic EEG data
def example_usage(eeg_signal):
    # Generate sample data (synthetic EEG)
    fs = 250  # Sample rate (Hz)
    duration = 3 # seconds
    t = np.linspace(0, duration, int(fs * duration), endpoint=False)
    
    
    # Create filtered signals dictionary
    filtered_signals = {}
    
    # Apply bandpass filtering for each frequency band
    for band_name, (low_freq, high_freq) in freq_bands.items():
        filtered_signals[band_name] = bandpass_filter(eeg_signal, low_freq, high_freq, fs)
    
    # Plot the original and filtered signals
    plt.figure(figsize=(12, 10))
    
    # Original signal
    plt.subplot(len(freq_bands) + 1, 1, 1)
    plt.plot(t, eeg_signal)
    plt.title('Original Signal')
    plt.ylabel('Amplitude')
    
    # Filtered signals
    for i, (band_name, filtered_signal) in enumerate(filtered_signals.items(), start=2):
        plt.subplot(len(freq_bands) + 1, 1, i)
        plt.plot(t, filtered_signal)
        plt.title(f'{band_name} ({freq_bands[band_name][0]}-{freq_bands[band_name][1]} Hz)')
        plt.ylabel('Amplitude')
    
    plt.xlabel('Time (s)')
    plt.tight_layout()
    plt.show()


example_usage(epochs_example[0][0][:750])
```

```python
def Full_bandpass(epochs_d):
    fs = 250  # Sample rate (Hz)
    duration = 3 # seconds
    t = np.linspace(0, duration, int(fs * duration), endpoint=False)
    
    
    # Create filtered signals dictionary
    filtered_signals = []
    eeg_filtered = []
    
    # Apply bandpass filtering for each frequency band
    for eeg_signal in epochs_d:
        for band_name, (low_freq, high_freq) in freq_bands.items():
            filtered_signals.append(bandpass_filter(eeg_signal[:750], low_freq, high_freq, fs))
        eeg_filtered.append(np.array(filtered_signals))
        filtered_signals = []
    
    print(np.array(eeg_filtered).shape)
    return np.array(eeg_filtered)[:,:,:,:750]

Full_bandpass(epochs_example)
```

```python
import torch 
import torch.nn as nn

class FCBNet(nn.Module):
    """
    Filter Bank Common Spatial Pattern Network for EEG representation learning
    """
    def __init__(self, eeg_channels, n_bandpass, m_filter, n_times, n_embedding, hidden_size):
        super().__init__()

        self.eeg_channels = eeg_channels
        self.n_bandpass = n_bandpass
        self.m_filter = m_filter
        self.n_times = n_times
        self.n_embedding = n_embedding
        self.hidden_size = hidden_size

        self.spacial_conv = nn.ModuleList([
            nn.Conv2d(1, self.m_filter, (self.eeg_channels,1), bias = False)
            for _ in range(self.n_bandpass)
        ])

        self.flatten_dim = self.n_bandpass * self.m_filter

        self.embeding = nn.Sequential(
            nn.Linear(self.flatten_dim, self.hidden_size),
            nn.BatchNorm1d(self.hidden_size),
            nn.ELU(),
            nn.Linear(self.hidden_size, self.n_embedding),
            nn.BatchNorm1d(self.n_embedding)
        )

    def forward(self, eeg_signal):
        """
            Given eeg_signal forward function returns embedding vector for this eeg_signal
        """
        batch_size = eeg_signal.shape[0]

        features = []
        for i in range(self.n_bandpass):
            help = self.spacial_conv[i](eeg_signal[:,i:i+1,:,:])

            help = help ** 2

            help = torch.mean(help, dim=3, keepdim=True)

            help = torch.log(help + 1e-6)

            help = help.view(batch_size,self.m_filter)

            features.append(help)
        
        features = torch.cat(features, dim=1)

        embeddings = self.embeding(features)

        return embeddings
    
```

```python
import random

def generate_triplets(data_dict, batch_size, type):
    """
    Generate triplets (anchor, positive, negative) from a dictionary dataset.

    Parameters:
    -----------
    data_dict : Dict[str, List[Any]]
        A dictionary where keys are participant labels (e.g., "participant_1") and
        values are lists of data samples for each participant.
    num_triplets : int, optional
        Number of triplets to generate. Default is 10000.

    Returns:
    --------
    List[Tuple[Any, Any, Any]]
        A list of triplets, where each triplet is (anchor, positive, negative)
    """
    all_participants = []
    if type == "Train":
        all_participants = [f"participant_{i}" for i in range(0,20)]
    elif type == "Valid":
        all_participants = [f"participant_{i}" for i in range(20,23)]
    elif type == "Test":
        all_participants = [f"participant_{i}" for i in range(23,26)]
    else:
        all_participants = [f"participant_{i}" for i in range(26,30)]

    triplets = []
    
    for _ in range(batch_size):
        # Randomly select an anchor participant
        anchor_participant = random.choice(all_participants)
        
        # Select a negative participant (different from anchor)
        negative_participants = [p for p in all_participants if p != anchor_participant]
        negative_participant = random.choice(negative_participants)
        
        # Get data samples for both participants
        anchor_samples = data_dict[anchor_participant]
        negative_samples = data_dict[negative_participant]
        
        # Make sure we have at least 2 samples from the anchor participant
        if len(anchor_samples) < 2:
            continue
        
        # Randomly select anchor and positive samples (from the same participant)
        idx1, idx2 = random.sample(range(len(anchor_samples)), 2)
        anchor = anchor_samples[idx1]
        positive = anchor_samples[idx2]
        
        # Randomly select a negative sample (from different participant)
        negative = random.choice(negative_samples)
        
        triplets.append(torch.stack([torch.tensor(anchor, dtype=torch.float), torch.tensor(positive, dtype=torch.float), torch.tensor(negative, dtype=torch.float)]))
    
    return torch.stack(triplets).transpose(0,1)

```
