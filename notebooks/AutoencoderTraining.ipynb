```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# Define frequency bands (corrected the typo in alpha2)
freq_bands = {
    "delta": (0.5, 4),
    "theta": (4, 8),
    "alpha1": (8, 10),
    "alpha2": (10, 12),
    "beta1": (12, 16),
    "beta2": (16, 20),
    "beta3": (20, 30),
    "gamma1": (30, 40),
    "gamma2": (40, 100),
}

def bandpass_filter(data, lowcut, highcut, fs, order=4):
    """
    Apply a bandpass filter to the signal.
    
    Parameters:
    - data: 1D array of signal data
    - lowcut: low frequency cutoff
    - highcut: high frequency cutoff
    - fs: sampling frequency (Hz)
    - order: filter order
    
    Returns:
    - filtered_data: bandpass filtered signal
    """
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    
    # Create Butterworth bandpass filter
    b, a = signal.butter(order, [low, high], btype='band')
    
    # Apply filter
    filtered_data = signal.filtfilt(b, a, data)
    
    return filtered_data

# Example usage with synthetic EEG data
def example_usage(eeg_signal):
    # Generate sample data (synthetic EEG)
    fs = 250  # Sample rate (Hz)
    duration = 3 # seconds
    t = np.linspace(0, duration, int(fs * duration), endpoint=False)
    
    
    # Create filtered signals dictionary
    filtered_signals = {}
    
    # Apply bandpass filtering for each frequency band
    for band_name, (low_freq, high_freq) in freq_bands.items():
        filtered_signals[band_name] = bandpass_filter(eeg_signal, low_freq, high_freq, fs)
    
    # Plot the original and filtered signals
    plt.figure(figsize=(12, 10))
    
    # Original signal
    plt.subplot(len(freq_bands) + 1, 1, 1)
    plt.plot(t, eeg_signal)
    plt.title('Original Signal')
    plt.ylabel('Amplitude')
    
    # Filtered signals
    for i, (band_name, filtered_signal) in enumerate(filtered_signals.items(), start=2):
        plt.subplot(len(freq_bands) + 1, 1, i)
        plt.plot(t, filtered_signal)
        plt.title(f'{band_name} ({freq_bands[band_name][0]}-{freq_bands[band_name][1]} Hz)')
        plt.ylabel('Amplitude')
    
    plt.xlabel('Time (s)')
    plt.tight_layout()
    plt.show()


example_usage(epochs_example[0][0][:750])
```

```python
def Full_bandpass(epochs_d):
    fs = 250  # Sample rate (Hz)
    duration = 3 # seconds
    t = np.linspace(0, duration, int(fs * duration), endpoint=False)
    
    
    # Create filtered signals dictionary
    filtered_signals = []
    eeg_filtered = []
    
    # Apply bandpass filtering for each frequency band
    for eeg_signal in epochs_d:
        for band_name, (low_freq, high_freq) in freq_bands.items():
            filtered_signals.append(bandpass_filter(eeg_signal[:750], low_freq, high_freq, fs))
        eeg_filtered.append(np.array(filtered_signals))
        filtered_signals = []
    
    print(np.array(eeg_filtered).shape)
    return np.array(eeg_filtered)[:,:,:,:750]

Full_bandpass(epochs_example)
```

```python
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
#https://www.sciencedirect.com/science/article/pii/S1053811924004439

torch.manual_seed(42)

class BandpassVAEEncoder(nn.Module):
    def __init__(self, latent_dim=32):
        super().__inti__()

        self.latent_dim = latent_dim

        self.encoder = nn.Sequential(
            #Input shape [batch, 1, 4, 750]
            nn.Conv2d(1, 16, kernel_size=(2,15), stride=(1,2), padding=(0,7)),
            nn.BatchNorm2d(16),
            nn.GELU(),
            #Shape [batch, 16, 3, 375]

            nn.Conv2d(16, 32, kernel_size=(2,7), stride=(1,2), padding=(0,3)),
            nn.BatchNorm2d(32),
            nn.GELU(),
            #Shape [batch, 32, 2, 188]

            nn.Conv2d(32, 64, kernel_size=(2,5), stride=(1,2), padding=(0,2)),
            nn.BatchNorm2d(64),
            nn.GELU(),
            #Shape [batch, 64, 1, 94]

            nn.Conv2d(64, 128, kernel_size=(1,3), stride=(1,2), padding=(0,1)),
            nn.BatchNorm2d(128),
            nn.GELU(),
            #Shape [batch, 128, 1, 47]

            nn.Flatten()
            #Shape [batch, 128 * 1 * 47]

        )

        self.feature_size = 128 * 1 * 47

        # Latent space parameters
        self.fc_mu = nn.Linear(self.feature_size, self.latent_dim)
        self.fc_log_var = nn.Linear(self.feature_size, self.latent_dim)

    def forward(self, x):
        # x shape: [batch,channels, lenght]
        # Add dimension for single filter
        if len(x.shape)==3:
            x = x.unsqueeze(1)
        
        features = self.encoder(x)

        mu = self.fc_mu(features)
        log_var = self.log_var(features)
        
        return mu, log_var, features
```

```python
class BandpassVAEDencoder(nn.Module):
    def __init__(self, latent_dim=32, feature_size=128*1*47):
        super().__init__()

        self.latent_dim = latent_dim
        self.feature_size = feature_size

        self.latent_to_features = nn.Sequential(
            nn.Linear(self.latent_dim, self.feature_size),
            nn.GELU()
        )
        
        self.decoder = nn.Sequential(
            #Start with [batch, 128, 1, 47]
            nn.ConvTranspose2d(64, 128, kernel_size=(1,3), stride=(1,2), padding=(0,1), output_padding=(0, 1)),
            nn.BatchNorm2d(128),
            nn.GELU(),
            #Shape [batch, 64, 1, 94]

            nn.ConvTranspose2d(32, 64, kernel_size=(2,5), stride=(1,2), padding=(0,2), output_padding=(0, 1)),
            nn.BatchNorm2d(64),
            nn.GELU(),
            #Shape [batch, 32, 2, 188]

            nn.ConvTranspose2d(16, 32, kernel_size=(2,7), stride=(1,2), padding=(0,3), output_padding=(0, 1)),
            nn.BatchNorm2d(32),
            nn.GELU(),
            #Shape [batch, 16, 3, 375]

            nn.ConvTranspose2d(1, 16, kernel_size=(2,15), stride=(1,2), padding=(0,7), output_padding=(0, 1)),
            nn.BatchNorm2d(16),
            nn.GELU(),
            #Final shape [batch, 1, 4, 750]
        )

    def forward(self, z):
        features = self.latent_to_features(z)
        features = features.view(-1,128,1,47)

        output = self.decoder(features)

        return output.squeeze(1)
```
