# Przykadowy sygna dla jednego pomiaru

```python
sample = eeg_df['epoch'].iloc[0]
sample.shape
```

```python
sample = eeg_df['epoch'].iloc[0][0]
```

```python
plt.figure(figsize=(15, 7))
plt.plot(sample[0])
plt.xlabel('Time')
plt.title('Przykadowy sygna dla kanau 1')
plt.ylabel('Amplitude')
plt.show()
```

```python
plt.figure(figsize=(15, 7))
for channel in sample:
    plt.plot(channel, alpha=0.5)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
plt.title(f'Przykadowy sygna dla wszsytkich kanaw')
plt.legend(['Kana 1', 'Kana 2', 'Kana 3', 'Kana 4'])
plt.show()
```

```python
plt.figure(figsize=(15, 7))
for signal in eeg_df['epoch'].iloc[0]:
    plt.plot(signal[0], alpha=0.5)
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
plt.title(f'Przykadowy sygna dla pierwszego kanau dla wszsytkich prbek')
plt.show()
```
