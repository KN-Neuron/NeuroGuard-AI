```python
import numpy as np
freq_bands = {
    "delta": (1, 4),
    "theta": (4, 8),
    "alpha": (8, 14),
    "beta": (14, 31),
    "gamma": (31, 49),
}
spectrum = epochs.compute_psd(picks="eeg", fmin=0.0, fmax=49.0)
psds, freqs = spectrum.get_data(return_freqs=True)

psds /= np.sum(psds, axis=-1, keepdims=True)
X = []
for fmin, fmax in freq_bands.values():
    psds_band = psds[:,:,(freqs >= fmin) & (freqs < fmax)].mean(axis=-1)
    X.append(psds_band.reshape(len(psds), -1))

data = np.concatenate(X,axis=1)
```

```python
y = []
for i in range(len(epochs)):
    y.append(epochs[i].events[:,-1])
```

```python
import torch
```

```python
def get_data(X_train, y_train):
    # Example: replace these with your actual data
    # y_train: tensor of shape (N,) containing class labels
    # X_train: tensor of shape (N, features)

    # Step 1: Count occurrences per class
    unique_classes, counts = torch.unique(y_train, return_counts=True)

    max_class = unique_classes[torch.argmax(counts)]

    # Step 3: Initialize list of selected indices
    selected_indices = []

    for cls, count in zip(unique_classes, counts):
        if cls == max_class:
            # randomly select 23 indices from largest class
            max_indices = (y_train == cls).nonzero(as_tuple=True)[0]
            chosen = max_indices[torch.randperm(len(max_indices))[:23]]
            selected_indices.append(chosen)
        else:
            # include all indices for this class
            class_indices = (y_train == cls).nonzero(as_tuple=True)[0]
            selected_indices.append(class_indices)

    # Step 4: Concatenate all selected indices
    final_indices = torch.cat(selected_indices)

    # Step 5: Index into X_train and y_train
    X_selected = X_train[final_indices]
    y_selected = y_train[final_indices]

    # Shuffle the selected data
    perm = torch.randperm(len(final_indices))

    X_selected = X_selected[perm]
    y_selected = y_selected[perm]

    
    return X_selected,y_selected
```

```python
y_train = torch.tensor(y[:int(0.8*len(y))],dtype=torch.float)
y_test = torch.tensor(y[int(0.8*len(y)):],dtype=torch.float)
X_train = torch.tensor(data[:int(0.8*len(data))],dtype=torch.float)
X_test = torch.tensor(data[int(0.8*len(data)):],dtype=torch.float)
```

```python
X_train.shape, X_test.shape,y_train.shape,y_test.shape
```

```python
import torch.nn as nn
NUM_CLASSES = 6
EPOCHS = 100
INPUT_DIM = 20
HIDDEN_DIM = 128
```

```python
class Simple_MLP(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super().__init__()
        self.f1 = nn.Linear(input_dim, hidden_dim)
        self.f2 = nn.Linear(hidden_dim,hidden_dim)
        self.f3 = nn.Linear(hidden_dim,output_dim)
        self.gelu = nn.GELU()
    
    def forward(self, x):
        return self.f3(self.gelu(self.f2(self.gelu(self.f1(x)))))
```

```python
model = Simple_MLP(INPUT_DIM,HIDDEN_DIM,NUM_CLASSES)
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.0001)
```

```python
y_train.shape[0]
```

```python
flat = y_train.flatten().to(torch.long)
#print(flat)
one_hot = torch.zeros((y_train.shape[0], NUM_CLASSES))
for i in range(y_train.shape[0]):
    one_hot[i][flat[i]-1]=1
```

```python
max_test_acc = 0
for epoch in range(EPOCHS*1000):
    model.train()
    X_tr, y_tr = get_data(X_train,y_train)

    flat = y_tr.flatten().to(torch.long)
    one_hot = torch.zeros((y_tr.shape[0], NUM_CLASSES))
    for i in range(y_tr.shape[0]):
        one_hot[i][flat[i]-1]=1

    pred = model(X_tr)
    
    loss = loss_fn(pred,one_hot)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    ile = 0
    predtest = model(X_test)
    for i in range(len(y_test)):
        ile += predtest[i].argmax() == y_test[i]-1

    test_acc = ile/len(y_test)
    if(test_acc>max_test_acc):
        max_test_acc=test_acc
        torch.save(model.state_dict(), f"models/checkpoint{epoch}")
    print("Epoch: ", epoch, "Loss: ", loss.item(),"acc: ", test_acc,"max: ", max_test_acc)
```
